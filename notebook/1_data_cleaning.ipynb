## Data Cleaning
In this notebook, I will do data-cleaning from the raw data from here: https://communitynotes.x.com/guide/en/under-the-hood/download-data 

```python
import pandas as pd
import zipfile
import numpy as np 
```

```python
zip_path = '../data/raw/noteStatusHistory-00000.zip'
with zipfile.ZipFile(zip_path, 'r') as zip_ref:

  tsv_name = zip_ref.namelist()[0]
  with zip_ref.open(tsv_name) as tsv_file:
    df_nsh = pd.read_csv(tsv_file, sep='\t')

df_nsh.head()
```

```python
zip_path = '../data/raw/notes-00000.zip'
with zipfile.ZipFile(zip_path, 'r') as zip_ref:

  tsv_name = zip_ref.namelist()[0]
  with zip_ref.open(tsv_name) as tsv_file:
    df_nd = pd.read_csv(tsv_file, sep='\t')

df_nd.head()
```

### Merge two dataframes on "note ID"

```python
df_merged = pd.merge(df_nsh, df_nd, on=['noteId','createdAtMillis'])

df_merged = df_merged[["summary", "currentStatus", "createdAtMillis"]]
df_merged.head()
```

```python
df = df_merged
```

### Clean the data to meet the following conditions
- "currentStatus" is either "CURRENTLY_RATED_HELPFUL" or "CURRENTLY_RATED_NOT_HELPFUL"
- Notes posted from 2024-01-01 to 2024-12-31
- "summary" includes "Trump" or "Biden" or "Kamala" or "Harris
- Notes in English

```python
df['currentStatus'] = df["currentStatus"].replace('NEEDS_MORE_RATINGS', np.nan)
df = df.dropna(subset = "currentStatus")

df.head()
```

```python
import datetime

# change unix into datetime
df["datetime"] = df["createdAtMillis"]
for i in range(len(df)):
    df["datetime"].iloc[i] = datetime.datetime.fromtimestamp(df["createdAtMillis"].iloc[i] / 1000)
    df["datetime"].iloc[i] = df["datetime"].iloc[i].strftime("%Y-%m-%d")
    
df = df[(df["datetime"] >= "2024-01-01") & (df["datetime"] <= "2024-12-31")]
df.head()
```

```python
df = df[df["summary"].str.contains(r"Kamala|Harris|Donald Trump|Trump|Joe Biden|Biden", case=False, na=False)]
df.head(5)
```

```python
import langid

df = df[df["summary"].apply(lambda x: langid.classify(x)[0] == "en")]
```

```python
# download the file
df.to_csv('../data/processed/notes_2024.csv', index=False)
```

### Extract FQDN

```python
import re

def extract_fqdn_list(text):
    if pd.isna(text):
        return []
    pattern = r"https://([a-zA-Z0-9.-]+)"
    domains = re.findall(pattern, str(text))
    return domains

df['fqdn'] = df['summary'].apply(extract_fqdn_list)

df.head()
```

```python
df = df.reset_index()
df = df.drop(columns=['index'])
df.head()
```
